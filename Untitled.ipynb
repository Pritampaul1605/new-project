```python
import re
from collections import Counter
from math import log

# Basic tokenizer
def tokenize_text(text):
    # Remove punctuation, lower case, and split sentences
    sentences = re.split(r'(?<!\w\.\w.)(?<![A-Z][a-z]\.)(?<=\.|\?)\s', text)
    words = re.findall(r'\w+', text.lower())
    return sentences, words

# Removing common stopwords
def remove_stopwords(words, stopwords):
    return [word for word in words if word not in stopwords]

# Stemming (we can add more complex lemmatization if needed)
def stem_word(word):
    suffixes = ['ing', 'ly', 'ed', 'ious', 'ies', 'ive', 'es', 's', 'ment']
    for suffix in suffixes:
        if word.endswith(suffix):
            return word[:-len(suffix)]
    return word
```

```python
def frequency_based_summarization(text, stopwords, summary_ratio=0.2):
    sentences, words = tokenize_text(text)
    words = remove_stopwords(words, stopwords)
    
    # Frequency of each word
    word_frequencies = Counter(words)
    print(word_frequencies)
    
    # Scoring sentences
    sentence_scores = {}
    for i, sentence in enumerate(sentences):
        sentence_words = re.findall(r'\w+', sentence.lower())
        for word in sentence_words:
            word = stem_word(word)
            if word in word_frequencies:
                if i not in sentence_scores:
                    sentence_scores[i] = word_frequencies[word]
                else:
                    sentence_scores[i] += word_frequencies[word]
    
    # Sort sentences by score
    summary_length = int(len(sentences) * summary_ratio)
    print(summary_length)
    summary_sentences = sorted(sentence_scores, key=sentence_scores.get, reverse=True)[:summary_length]
    print(summary_sentences)
    
    # Return the best scoring sentences as a summary
    return ' '.join([sentences[i] for i in summary_sentences])
```

```python
def compute_tf(word_freqs):
    tf = {}
    max_freq = max(word_freqs.values())
    for word, count in word_freqs.items():
        tf[word] = count / max_freq
    return tf

def compute_idf(documents):
    idf = {}
    total_docs = len(documents)
    for document in documents:
        for word in set(document):
            idf[word] = idf.get(word, 0) + 1
    
    for word, count in idf.items():
        idf[word] = log(total_docs / (1 + count))
    
    return idf

def extract_keywords(text, stopwords, corpus, top_n=5):
    sentences, words = tokenize_text(text)
    words = remove_stopwords(words, stopwords)
    
    # Frequency of words in the document
    word_frequencies = Counter(words)
    
    # TF and IDF
    tf = compute_tf(word_frequencies)
    idf = compute_idf(corpus)  # Assume 'corpus' is a list of tokenized documents
    
    # Calculate TF-IDF
    tf_idf = {word: tf[word] * idf.get(word, 0) for word in words}
    
    # Get top N keywords
    top_keywords = sorted(tf_idf, key=tf_idf.get, reverse=True)[:top_n]
    
    return top_keywords
```

```python
def adjust_based_on_length(text, stopwords, corpus):
    sentences, words = tokenize_text(text)
    length = len(sentences)
    
    # For short documents, summarize more briefly
    if length < 5:
        summary = frequency_based_summarization(text, stopwords, summary_ratio=0.3)
        keywords = extract_keywords(text, stopwords, corpus, top_n=3)
    # For medium-length documents
    elif length < 15:
        summary = frequency_based_summarization(text, stopwords, summary_ratio=0.2)
        keywords = extract_keywords(text, stopwords, corpus, top_n=5)
    # For long documents
    else:
        summary = frequency_based_summarization(text, stopwords, summary_ratio=0.1)
        keywords = extract_keywords(text, stopwords, corpus, top_n=10)
    
    return summary, keywords
```

```python
import PyPDF2
from PyPDF2.errors import PdfReadError

def read_pdf_file(file_path):
    try:
        # Attempt to open the file in binary mode
        with open(file_path, 'rb') as file:
            reader = PyPDF2.PdfReader(file)

            # Check for an empty or invalid PDF
            if not reader.pages:
                raise ValueError("The PDF file is empty or corrupted.")

            # Read the first page as an example
#             page = reader.pages[0]
#             text = page.extract_text()
            page = len(reader.pages)
            pdftext = []
            for i in range(page) :
                pdftext.append(reader.pages[i].extract_text())
            return pdftext
#             print("Text extracted from the first page:\n", text)

    except FileNotFoundError:
        print("The specified PDF file was not found. Please check the file path.")
    except PdfReadError:
        print("The PDF file could not be read. It may be corrupted or an unsupported format.")
    except IOError:
        print("An I/O error occurred while accessing the file.")
    except ValueError as ve:
        print(f"Error processing the file: {ve}")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")


```
